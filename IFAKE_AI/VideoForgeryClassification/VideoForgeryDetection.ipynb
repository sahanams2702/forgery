```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
from keras.models import load_model
```

```python
model = load_model("ResNet50_Model/forgery_model.hdf5")
```

```python
model.summary()
```

```python
vid_src = "C:/Users/User/ML/Video_Forgery_Detection/video_tampering_dataset/testing/sample2.mp4" #videos/Original/05_original.mp4"
vid = []

sumframes =0
cap= cv2.VideoCapture(vid_src)
#cap.set(3,240)
#cap.set(4,320)

while(cap.isOpened()):
    ret, frame = cap.read()
    #compressImage = cv2.resize(frame, (0, 0), fx = 0.1, fy = 0.1)
    
    

    if ret == False:
        break
    
    b = cv2.resize(frame,(320,240),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)

    sumframes +=1
    
    vid.append(b)
cap.release()
    
print("No. Of Frames in the Video: ",sumframes)
```

```python
Xtest = np.array(vid)
print(Xtest.shape)
```

```python
output = model.predict(Xtest)
```

```python
print(output.shape)
```

```python
output = output.reshape((-1))
print(output.shape)
print(output)
```

```python
results = []
for i in output:
    if i>0.5:
        results.append(1)
    else:
        results.append(0)
```

```python
print(len(results))
print(results)
```

```python
forge_flag = 0
for i in results:
    if i == 1:
        forge_flag = 1
        break
        
if forge_flag == 0:
    print("The video is not forged")
    
else:
    print("The video is forged")
    print("Number of Forged Frames in the video: ",sum(results))
```
